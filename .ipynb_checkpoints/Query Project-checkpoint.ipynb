## Part 1: Commuter Trips

I will start with some exploratory analytics and visualizations to help define what I consider to be a commuter trip.

```python
%%bigquery start_times --use_rest_api

SELECT 
  EXTRACT(HOUR FROM start_date) as hours,
  COUNT(*) as trips
FROM `bigquery-public-data.san_francisco.bikeshare_trips`
WHERE EXTRACT(DAYOFWEEK FROM start_date) NOT IN (1,7)
GROUP BY 1
ORDER BY 1 ASC
```

```python
%%bigquery trip_duration --use_rest_api

SELECT 
    ROUND(duration_sec/360.0, 0) as hr_round,
    COUNT(*) as trips
FROM `bigquery-public-data.san_francisco.bikeshare_trips`
WHERE EXTRACT(DAYOFWEEK FROM start_date) NOT IN (1,7)
    AND ROUND(duration_sec/360.0, 0) < 24
GROUP BY 1
ORDER BY 1 ASC
```

```python
fig, (ax1, ax2) = plt.subplots(1,2)

fig.suptitle("Weekday Trips")

ax1.bar('hours','trips', data = start_times)
ax1.set_xlabel("Trip Start Time (Hour)")
ax1.set_ylabel("Trips")
ax1.set_title("Start Time Distribution")
for x in (6,10,15,19):
    ax1.axvline(x = x, c = 'gray', linestyle = 'dashed')
    
ax2.bar('hr_round','trips', data = trip_duration)
ax2.set_xlabel("Trip Duration (Hours)")
ax2.set_title("Duration Distribution")

fig.set_size_inches(15,7)
fig.tight_layout()
```

**Start Time Distribution:** Looking at the first plot on the left, I decided to set the time range for commute hours to be weekdays 6am - 10am and 3pm - 7pm. The two four-hour-long ranges encompass the two distinct peaks we see in the first bar plot.   

**Duration Distribution:** Nearly all of the trips taken on the weekdays are finished within a few hours. The plot only shows the trips taken within 24 hrs since trip durations exceeding that are small in number and also not relevant to defining the commuter trips.

### What are the 5 most popular trips that you would call "commuter trips"?

Using the insights we gained above, I defined the conditions I will apply to determine what makes a commuter trip:
- different start and end station
- starts during defined weekday commute hours (6am -10am, 3pm - 7pm)
- duration of the trip is within 5 minutes of the minimum time recorded to make that commute

```python
%%bigquery commuter --use_rest_api

WITH min AS (
    SELECT
        start_station_name,
        end_station_name,
        MIN(ROUND(duration_sec / 60.0, 0)) AS min_time
    FROM `bigquery-public-data.san_francisco.bikeshare_trips`
    GROUP BY 1,2 
)
SELECT
    b.start_station_name,
    b.end_station_name,
    COUNT(*) AS trips
FROM `bigquery-public-data.san_francisco.bikeshare_trips` AS b
LEFT JOIN min m
    ON b.start_station_name = m.start_station_name
    AND b.end_station_name = m.end_station_name
WHERE EXTRACT(HOUR FROM start_date) IN (6,7,8,9,10,15,16,17,18,19)
    AND EXTRACT(DAYOFWEEK FROM start_date) NOT IN (1,7)
    AND b.start_station_name <> b.end_station_name
    AND ROUND(b.duration_sec / 60.0, 0) <= m.min_time + 5.0
GROUP BY 1,2
ORDER BY 3 DESC
LIMIT 5
```

### 2.1 Bike Availability 

```python
%%bigquery no_bike --use_rest_api

SELECT
    EXTRACT(DAYOFWEEK FROM time) as day,
    EXTRACT(HOUR FROM time) as hour,
    COUNT(DISTINCT station_id) as stations,
    COUNT(*) as instances
FROM `bigquery-public-data.san_francisco.bikeshare_status`
WHERE bikes_available = 0
GROUP BY 1,2
ORDER BY 1,2
```

This plot shows the instances of no bikes being available at any given station throughout the week. There are two distinct peaks on the weekdays that translate roughly to the aforementioned commute hours. It is concerning how there are upwards 15000 cases of stations not having any bikes available during the busiest hours. We risk deterring regular commuters and potential users due to the unreliable supply of bikes especially on the commute home. 

```python
%%bigquery bike_docks --use_rest_api

WITH temp AS(
SELECT
    station_id,
    CASE WHEN bikes_available = 0 THEN 1 ELSE 0 END AS no_bike, 
    CASE WHEN docks_available = 0 THEN 1 ELSE 0 END AS no_dock,
FROM `bigquery-public-data.san_francisco.bikeshare_status` a
)
SELECT
    a.name,
    a.dockcount,
    SUM(t.no_bike) as no_bikes,
    SUM(t.no_dock) as no_docks,
    ROUND(SUM(t.no_bike) / COUNT(t.station_id) * 100.0, 2) as percent_bike
FROM `bigquery-public-data.san_francisco.bikeshare_stations` a
RIGHT JOIN temp as t
ON a.station_id = t.station_id
GROUP BY 1,2
ORDER BY 5 DESC
LIMIT 11
```

```python
bike_docks[1:]
```

Grouping by stations, these are the counts of instances with no bikes available and no docks available, ranked by the percent of instances where there are no bikes. Not having any docks is also an issue but it is less urgent since users have already taken the trip and can find another station nearby. Not having any bikes available is a greater deterrent for potential users. It is worthwhile to potentially **increase the number of docks at specific stations** especially if there are very few instances of no docks being available.

### 2.2 Target Non-Subscription Customers

```python
%%bigquery sub_days --use_rest_api

SELECT 
    subscriber_type,
    CASE WHEN (duration_sec / (24 * 360.0) < 10) THEN CEILING(duration_sec / (24 * 360.0)) ELSE 10 END as day,
    COUNT(*) as total_trips
FROM `bigquery-public-data.san_francisco.bikeshare_trips`
GROUP BY 1,2
ORDER BY 2 ASC
```

```python
sub = pd.pivot_table(sub_days, index = 'day', values = 'total_trips', columns = 'subscriber_type')
sub
```

*Note: any trips that take longer than 10 days to complete are aggregated in the last 10.0 day row.*

Observing differences between subscribers and customers, we can see that the bulk of rides are completed within a day and subscribers make up most of the trips taken already. However, we can see that customers have relatively much more occurances of taking a bike for longer periods of time. For whatever reason (inability to return the bike, taking it for longer travels, etc.), they hold onto the bike which limits other people's use of the bike (further contributes to the poor bike availability).   

To help combat this issue, I would suggest an **extra perk for subscription members that encourage returning the bikes within a day**. For example, if you are a **new monthly/annual subscriber you can get a 5% discount on next month's rides if you finish x amount of rides within the day/hour**. This will encourage non-subscription users to subscribe and return their bike in a timely manner. Since the bulk of trips are subscriber short-term commuter type rides, it is not very efficient to continuously target existing subscribers or commuters.

```python
%%bigquery sub_stats --use_rest_api

SELECT 
    subscriber_type,
    COUNT(*) as trips,
    ROUND(AVG(duration_sec / 60.0),2) as avg_mins,
    ROUND(STDDEV(duration_sec / 60.0),2) as sd_mins,
    SUM(CASE WHEN (duration_sec / 60.0) < 5 THEN 1 ELSE 0 END) as within_5mins,
    SUM(CASE WHEN (duration_sec / 360.0) < 1 THEN 1 ELSE 0 END) as within_1hr,
    SUM(CASE WHEN (duration_sec / (24*360.0)) < 1 THEN 1 ELSE 0 END) as within_1day,
    SUM(CASE WHEN (duration_sec / (24*360.0)) >= 1 THEN 1 ELSE 0 END) as greater_1day,
    SUM(CASE WHEN (duration_sec / (7*24*360.0)) >= 1 THEN 1 ELSE 0 END) as greater_1week,
    SUM(CASE WHEN start_station_name = end_station_name AND (duration_sec / 60.0) < 5 THEN 1 ELSE 0 END) as no_ride, 
    SUM(CASE WHEN start_station_name = end_station_name THEN 1 ELSE 0 END) as start_end,
    ROUND(AVG(CASE WHEN start_station_name = end_station_name THEN duration_sec / 60.0 ELSE NULL END),2) as start_end_avg_mins
FROM `bigquery-public-data.san_francisco.bikeshare_trips`
GROUP BY 1
ORDER BY 1,2 DESC
```

```python
sub_stats
```

Further summary statistics comparing the customer and subscriber cohorts emphasizes more areas of potential for non-subscribers.    

The average subscriber finishes their trip within 10 minutes and essentially all of them, **99.9%**, complete the trip within the day. Cases of rides with the same start and end is very low so we can reasonably say that the low average is largely due in part to subscribers making up the bulk of commuter rides. There is **no need to further target the subscriber cohort since they are actively engaging regularly as part of a commute.**    

If you look at the average non-subscription customer we see that their trips average an hour and 9% of their trips exceed a day. While the instances of accidental rides (puts the bike back within 5 minutes) is negligible across both groups, for customers **16%** of their rides end and start at the same station. These same start and end rides also average a little over 2 hours. This can possibly imply that many non-subscription riders are recreational. We should further **target and promote day passes to non-subscription members**, potentially with perks if they finish at the same station.
